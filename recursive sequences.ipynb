```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
def recurisive_sq(s1,w,itr ):
    sequence = [s1]
    si_minus_1=s1
    for i in range(itr):
        s_i = w*si_minus_1-w*(si_minus_1)**2
        sequence.append(s_i)
        si_minus_1 = s_i
    return sequence
```

```python
y = recurisive_sq(0.5,1,20)
t1=np.arange(0,21,1)


plt.plot(t1, y, 'bo', t1, y, 'k')
plt.title('recursive funtion with w=0.5 s=1')
plt.show()
```

```python
y = recurisive_sq(0.5,3,20)
t1=range(21)
plt.plot(t1, y, 'bo', t1, y, 'k')
plt.title('recursive funtion with w=0.5 s=3')
plt.show()
```
